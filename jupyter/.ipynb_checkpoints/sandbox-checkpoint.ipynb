```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(3, 3); # 3 qubits and 3 clasicall bits
qc.draw(output='mpl')
```

```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(3, 3); # 3 qubits and 3 clasicall bits
qc.measure([0, 1, 2], [0, 1, 2])
qc.draw(output='mpl')
```

```python
from qiskit.providers.aer import AerSimulator
sim = AerSimulator()

job = sim.run(qc)
result = job.result()
result.get_counts()
```
